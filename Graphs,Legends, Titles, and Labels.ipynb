```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [5,7,4]

x2 = [1,2,3]
y2 = [10,14,12]
plt.plot(x, y, label='First Line')
plt.plot(x2, y2, label='Second Line')
plt.xlabel('Plot Number')
plt.ylabel('Important var')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x=[1,2,3]
y=[4,5,6]

x2 = [7,8,2]
y2 = [1,11,14]


print(len(x))
print(len(y))

# can plot specifically, after just showing the defaults:
plt.plot(x,y)
plt.plot(x2,y2)

plt.title('Epic chart')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

# can plot specifically, after just showing the defaults:
plt.plot(x,y,'g', linewidth=5)
plt.plot(x2,y2,'c',linewidth=5)

plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

# can plot specifically, after just showing the defaults:
plt.plot(x,y,'g',linewidth=5, label='line one')
plt.plot(x2,y2,'c', linewidth=5, label='line two')

plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()

plt.show()
```
