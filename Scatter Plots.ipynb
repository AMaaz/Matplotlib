```python
import matplotlib.pyplot as plt

# Used to compare variables(This case in 2-dim)

x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y, label='skitscat', color='k', s=55, marker="o")
# for more markers, check out this website
#https://matplotlib.org/api/markers_api.html

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```
